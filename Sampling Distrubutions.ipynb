```python
import numpy as np
import scipy.stats as s
import matplotlib.pyplot as plt
```

1. Sampling Distribution of Mean 

when single dice throw

```python
dice_throws_samples = np.random.randint(low=1,high=7,size=(10000,10))
(dice_throws_samples)
```

We have n = 1000 samples each having 10 different variables from 1 to 6
Now, we calculate mean of each sample and then distribute this on graph which is called as sampling distribution of mean

```python
sample_means = np.mean(dice_throws_samples,axis = 1) # 1 means HORIZONTALL 
sample_means # mean of 1000 samples
plt.hist(sample_means)
```
